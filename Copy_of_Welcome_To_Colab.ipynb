```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("/content/employee_attrition_classification.csv")
df
```

```
df.info()
```

```
df["department"].unique()
```

```
df["department"].replace(['Finance', 'FINANCE', 'Finace', 'Financ', 'inance'], 'Finance', inplace=True)
df["department"].replace(['HR', 'Hr', 'H', 'R'], 'HR', inplace=True)
df["department"].replace(['Sales', 'SALES', 'Sals', 'Saes', 'ales'], 'Sales', inplace=True)
df["department"].replace(['Support', 'SUPPORT', 'Suppot', 'Suppor', 'Suport', 'Spport', 'upport'], 'Support', inplace=True)
df["department"].replace(['Engineering', 'ENGINEERING', 'Engineerig', 'Engneering'], 'Engineering', inplace=True)
df["department"].fillna(df["department"].mode()[0], inplace=True)
```

```
df["department"].unique()
```

```
df['department'].isna().count()
```

```
df["work_accident"].unique()
```

```
df["work_accident"].replace(['Yes'], 'Yes', inplace=True)
df["work_accident"].replace(['No', 'o', 'N', 'NO'], 'No', inplace=True)
df["work_accident"].fillna(df["work_accident"].mode()[0], inplace=True)
```

```
df["work_accident"].unique()
```

```
df['work_accident'].isna().count()
```

```
df['left_company'].unique()
```

```
df["left_company"].fillna(df["left_company"].mode()[0], inplace=True)
```

```
df['left_company'].unique()
```

```
df['left_company'].isna().count()
```

```
df['salary_k'].unique()
```

```
df["salary_k"].fillna(df["salary_k"].median(), inplace=True)
```

```
df['salary_k'].isna().count()
```

```
df["education_level"].unique()
```

```
df["education_level"].replace(['BACHELORS', 'Bachelors', 'Bacelors', 'Bachlors', 'Bachelos', 'Bahelors', 'Bacheors', 'achelors'], 'Bachelors', inplace=True)
df["education_level"].replace(['MASTERS', 'Masters', 'Msters', 'Maters', 'Masers'], 'Masters', inplace=True)
df["education_level"].replace(['PhD', 'Phd', 'PHD', 'Ph', 'hD', 'PD'], 'PhD', inplace=True)
df["education_level"].replace(['HIGH SCHOOL', 'High School', 'High school', 'HighSchool', 'High chool', 'High Scool', 'High Schol', 'Hig School', 'igh School'], 'High School', inplace=True)
df["education_level"].fillna(df["education_level"].mode()[0], inplace=True)
```

```
df["education_level"].unique()
```

```
df["education_level"].isna().count()
```

```
df["promotion_last_5years"].unique()
```

```
df["promotion_last_5years"].replace(['NO', 'N', 'No', 'o'], 'No', inplace=True)
df["promotion_last_5years"].replace(['YES', 'Ye', 'Yes', 'es'], 'Yes', inplace=True)
df["promotion_last_5years"].fillna('No', inplace=True)
```

```
df["promotion_last_5years"].unique()
```

```
df["promotion_last_5years"].isna().count()
```

```
df['years_at_company']=df['years_at_company'].fillna(df['years_at_company'].mean())
```

```
df['years_at_company'].isna().count()
```

```
df['salary_k_dup']=df['salary_k_dup'].fillna(df['salary_k_dup'].mean())
```

```
df['salary_k_dup'].isna().count()
```

```
df["last_performance_score"]=df["last_performance_score"].fillna(df["last_performance_score"].mean())
```

```
df["last_performance_score"].isna().count()
```

```
df = df.dropna(subset=['emp_id'])
```

```
df.info()
```

```
df.head()
```

```
df.to_csv("clean_employee.csv")
```
